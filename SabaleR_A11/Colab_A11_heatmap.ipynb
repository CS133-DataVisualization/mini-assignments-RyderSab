```python
sns.set_theme()
sns.catplot(data = ign_scores.sort_values(by = "Racing"),
            x = "Racing",
            y = "Platform",
            kind = "bar")
plt.title("Average IGN Rating for Racing Games on each Platform", wrap = True)
```

## Problem 5 - (4 points)
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
pokemon_data = pd.read_csv("https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv")
pokemon_data.head()
```

```python
pokemon_subset = pokemon_data[(pokemon_data["Type 1"] == "Water") | (pokemon_data["Type 1"] == "Normal") | (pokemon_data["Type 1"] == "Grass") | (pokemon_data["Type 1"] == "Bug") | (pokemon_data["Type 1"] == "Psychic")]
pokemon_subset = pokemon_subset.set_index("Name")

type_color = pokemon_subset["Type 1"].replace({"Water":"blue", "Normal":"gray", "Grass":"green", "Bug":"Yellow", "Psychic":"purple"})

pokemon_subset = pokemon_subset[["HP", "Attack", "Defense",	"Sp. Atk", "Sp. Def", "Speed"]]

sns.clustermap(data = pokemon_subset,
               row_colors = type_color,
               col_cluster = False,
               cmap = "viridis")
```
